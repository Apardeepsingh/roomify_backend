```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MultiLabelBinarizer
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import TfidfVectorizer
from sentence_transformers import SentenceTransformer
```

# Loading Data

```python
df = pd.read_json("catalog.json")

# keep only products that have all 3 dimensions (fit needs them)
df = df.dropna(subset=["width_mm", "depth_mm", "height_mm"]).reset_index(drop=True)
print(f"{len(df)} products with full dimensions")
df.head()
```

# Fit Cehck

```python
def fits_room(df, room_w, room_d, room_h):
    """Return products that physically fit a room (all sizes in mm).
    Checks both rotations of the footprint."""
    height_ok = df["height_mm"] <= room_h

    # normal orientation
    fit_a = (df["width_mm"] <= room_w) & (df["depth_mm"] <= room_d)
    # rotated 90 degrees
    fit_b = (df["width_mm"] <= room_d) & (df["depth_mm"] <= room_w)

    footprint_ok = fit_a | fit_b
    return df[height_ok & footprint_ok].copy()
```

```python
def available_floor_space(room_w, room_d, occupied_items=None, usable_fraction=0.6):
    """How much floor space is actually free for new furniture (in mm²).
    occupied_items: dataframe of items already in the room (or None if empty)."""
    total_floor = room_w * room_d * usable_fraction
    if occupied_items is None or len(occupied_items) == 0:
        return total_floor
    used = (occupied_items["width_mm"] * occupied_items["depth_mm"]).sum()
    return max(0, total_floor - used)


def candidates_for_room(catalog_df, room_w, room_d, room_h, occupied_items=None, room_type=None):
    """Stage 1: products that fit the room AND (optionally) match the room type."""

    df = catalog_df

    # NEW: if a room type is given, keep only products tagged for that room
    if room_type is not None:
        df = df[df["room_types"].apply(lambda rooms: room_type in rooms)]

    # existing geometry filters
    fitting = fits_room(df, room_w, room_d, room_h)

    free = available_floor_space(room_w, room_d, occupied_items)
    footprint = fitting["width_mm"] * fitting["depth_mm"]
    return fitting[footprint <= free].copy()
```

```python
# example:
# room with a sofa and a bed already inside
occupied = df.sample(2, random_state=2)
print("Already in the room:")
print(occupied[["name", "category"]])

candidates = candidates_for_room(df, 3000, 2500, 2400, occupied_items=occupied)
print(f"\n{len(candidates)} candidate products fit the leftover space")
candidates["category"].value_counts()
```

# Training Recommendation system

```python
# essentials a room usually has
ESSENTIAL = {"bed", "sofa", "table", "desk", "chair", "shelf"}
```

```python
def need_scores(candidates, user_items):
    """1 if user is missing this essential category,
       0.3 if non-essential and missing, 0 if user already has it."""
    have = set(user_items["category"].unique())

    def score(cat):
        if cat in have:
            return 0.0
        return 1.0 if cat in ESSENTIAL else 0.3

    return candidates["category"].map(score).values
```

```python

# load once, reuse
_model = SentenceTransformer("all-MiniLM-L6-v2")

def build_taste_vectors(user_items, candidates):
    """Use sentence embeddings on name + style + color.
    Category is excluded since the need score handles it."""

    def text_blob(df):
        return (
            df["name"].fillna("") + ". " +
            df["style"].fillna("") + ". " +
            df["color"].fillna("")
        ).tolist()

    user_vecs = _model.encode(text_blob(user_items), show_progress_bar=False)
    cand_vecs = _model.encode(text_blob(candidates), show_progress_bar=False)

    # average user items into one taste vector
    user_taste = user_vecs.mean(axis=0).reshape(1, -1)
    return user_taste, cand_vecs
```

```python
def taste_scores(user_taste, cand_vecs):
    """Cosine similarity, 0 to 1. Returns zeros if user has no tags at all."""
    if user_taste.sum() == 0:
        return np.zeros(cand_vecs.shape[0])
    sims = cosine_similarity(cand_vecs, user_taste).flatten()
    return np.clip(sims, 0, 1)
```

```python
def recommend(candidates, user_items, top_n=5, w_need=0.7, w_taste=0.3, use_taste=True):
    need = need_scores(candidates, user_items)

    out = candidates.copy()
    out["need"] = need

    if use_taste:
        user_taste, cand_vecs = build_taste_vectors(user_items, candidates)
        taste = taste_scores(user_taste, cand_vecs)
        out["taste"] = taste.round(3)
        out["score"] = (w_need * need + w_taste * taste).round(3)
    else:
        # no style matching: rank purely by need
        out["taste"] = None
        out["score"] = need

    return out.sort_values("score", ascending=False).head(top_n)
```

```python
# pretend user already has a bed and a sofa
user_items = df[df["category"].isin(["bed", "sofa"])].sample(2, random_state=4)
print("User has:")
print(user_items[["name", "category", "style", "color"]])

# get candidates that fit the room (from earlier)
room_candidates = candidates_for_room(df, 3500, 3000, 2400, occupied_items=user_items)

# recommend
top = recommend(room_candidates, user_items, top_n=5)
top[["name", "category", "style", "color", "need", "taste", "score"]]
```

```python
# user owns a bed, picks "office" as the room
user_items = df[df["category"] == "bed"].sample(1, random_state=4)
print("User owns:")
print(user_items[["name", "category"]])

# room-aware candidates
cands = candidates_for_room(df, 3500, 3000, 2400,
                            occupied_items=user_items, room_type="office")

top_with_taste = recommend(cands, user_items, top_n=5, use_taste=True)
print("\nWith taste (style matching ON):")
top_with_taste[["name", "category", "need", "taste", "score"]]
```

```python

top_no_taste = recommend(cands, user_items, top_n=5, use_taste=False)
print("Need only (style matching OFF):")
top_no_taste[["name", "category", "need", "taste", "score"]]
```
